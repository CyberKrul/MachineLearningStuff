# Linear Regression Using Python

## What is Regression?
* It is a technique for finding the relationship between two or more variables
* Change in dependent variable is associated with a change in one or more independent variables

---
* It is a technique that displays the relationship between variable 'y' based on the values of variable 'x'
* *eg* As the temperature drops people put on more jackets to keep warm

## Regression Use Cases
1. Temperature vs Number of Cones sold at ice-cream stores
2. Inches of rain vs newcars sold
3. Daily Snowfall vs number of skier visits

If you think there is a relationship between two events or variables, regression can help confirm it.

Therefore the system of equations of variables $m$ and $c$ are:

$c+m\mu_x = \mu_y$ 

$c\Sigma x+m\Sigma x^2 =\Sigma yx$

```python
# defining the data
import numpy as np
import matplotlib.pyplot as plt

#the data
x = np.array([[1,2,3,4,5]]).T
y = np.array([[4,3,4,2,5]]).T
meanx = np.mean(x)
meany = np.mean(y)

#plotting the data and their means
fig,ax = plt.subplots()
ax.plot(x,y,ls='',color='b',marker='o')
ax.plot(np.repeat(meanx,8),np.arange(0,8),ls='--',color='g',label=r'$\bar{x}$='+f'{meanx}')
ax.plot(np.arange(0,8),np.repeat(meany,8),ls='--',color='r',label=r'$\bar{y}$='+f'{meany}')
ax.set_xlabel('Independent Variable')
ax.set_ylabel('Dependent Variable')
ax.set_title('Understanding Linear Regression')
ax.legend()
```

```python
# solving the system of equations:
A = np.array([[1,meanx],[np.sum(x),np.sum(x**2)]])
B = np.array([[meany,np.sum(x*y)]]).T

c,m = np.linalg.solve(A,B)

#plotting the data and their means
fig,ax = plt.subplots()
ax.plot(x,y,ls='',color='b',marker='o')
ax.plot(np.arange(0,8),np.arange(0,8)*m+c,ls='--',color='g',label='Line of Best Fit')
ax.set_xlabel('Independent Variable')
ax.set_ylabel('Dependent Variable')
ax.set_title('Understanding Linear Regression')
ax.legend()
```

#### Evaluating the BFL
$R^2 = \Sigma(Predicted Dist-Mean)^2/\Sigma(Actual Dist-Mean)^2$

$R^2 = \Sigma(y_p-\bar{y})^2/\Sigma(y-\bar{y})^2$

```python
# calculating the R^2 score
num = np.sum((x*m+c-meany)**2)
denom = np.sum((y-meany)**2)
print(f'The performace of the model by the R^2 metric is {num/denom}')
```
